# Question 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
We need to find following details:
A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
#loading the webdrivers,URL & maximizing window

driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/")
driver.implicitly_wait(5)
```

```python
#click and open the search list

search=driver.find_element_by_id("searchButton")
search.click()
```

```python
#opting the particular search result

list_search=driver.find_element_by_xpath("//ul[@class='mw-search-results']/li[1]/div/a")
list_search.click()
```

```python
#creating an empty lists to store the scraping data-

Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
#finding all the details of most viewed videos on youtube by using xpath functionality-

#for ranking-
rank=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for r in rank:
    Rank.append(r.text)
    

#for name-
name=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for n in name:
    Name.append(n.text)
    
    
#for artist-
artist=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for a in artist:
    Artist.append(a.text)
    
    
#for upload date-
upload=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for u in upload:
    Upload_Date.append(u.text)

    
#for views-
views=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for v in views:
    Views.append(v.text)
```

```python
#now creating the dataframe from the above scraped data-
YOUTUBE=pd.DataFrame({})
YOUTUBE['Rank']=Rank[0:30]
YOUTUBE['Name']=Name[0:30]
YOUTUBE['Artist']=Artist[0:30]
YOUTUBE['Upload_Date']=Upload_Date[0:30]
YOUTUBE['Views']=Views[0:30]
```

```python
#loading the webdrivers,URL & maximizing window

driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
driver.get("https://www.bcci.tv/.")
driver.implicitly_wait(5)
```

```python
#clicking on International dropdown-

driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]').click()
```

```python
#selecting Features dropdown-

driver.find_element_by_xpath('//a[@class="navigation__link navigation__link--in-drop-down"]').click()
```

```python
#creating an empty list for storing the scraped data-

Match_Title=[]
Series=[]
Place=[]
Date=[]
Time=[]
```

```python
#finding all the details of  Match-

#for match title
title = driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
for i in title:
    Match_Title.append(i.text)
    
#for series
series = driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]/span[2]')
for i in series:
    Series.append(i.text)
    
#for place
place = driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span')
for i in place:
    Place.append(i.text)
    
#for date & month
date = driver.find_elements_by_xpath('//div[@class="fixture__datetime desktop-only"]/div/span')
month = driver.find_elements_by_xpath('//div[@class="fixture__datetime desktop-only"]/div/div/span[1]')
for i in range(len(date)):
    Date.append(date[i].text+month[i].text)
    
#for time
time = driver.find_elements_by_xpath('//div[@class="fixture__full-date"]/div/span[2]')
for i in time:
    Time.append(i.text)
```

```python
#creating the dataframe from  the above  scraped data-

BCCI=pd.DataFrame({})
BCCI['Match Title']=Match_Title
BCCI['Series']=Series
BCCI['Place']=Place
BCCI['Date']=Date
BCCI['Time']=Time
```
